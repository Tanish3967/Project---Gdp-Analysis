```python
data = df_pr.values
for i in range(1,len(df_pr)):
  prev = data[i-1][3]
  curr = data[i][3]
  print(data[i][2],(curr - prev))
```

```python
data = df_pr.values
# print("Gdp compared with previous year")
for i in range(1,len(df_pr)):
  prev = data[i-1][3]
  curr = data[i][3]
  print(data[i][2],"=",round((((curr - prev)/prev)*100),2))
```

```python
data = df_pr.values
x = []
y = []
for i in range(1,len(df_pr)):
  prev = data[i-1][3]
  curr = data[i][3]
  x.append(data[i][2])
  y.append(round((((curr - prev)/prev)*100),2))

plt.plot(x, y)
plt.xlabel("Year")
plt.ylabel("GDP Growth (%)")
plt.title("GDP Growth Over Time")
plt.show()
```

```python
l = []
for i in df["Country Name"].unique():
  l.append(i)
```

```python
for j in l:
  df_pr = df[df["Country Name"] == j]
  data = df_pr.values
  y = [0]
  for i in range(1,len(df_pr)):
    prev = data[i-1][3]
    curr = data[i][3]
    y.append(round((((curr - prev)/prev)*100),2))
  print(y)
  df_pr = df_pr.assign(GDP = y) # Overwrite df_pr with the new DataFrame that contains the 'GDP' column
  df_pr.plot(kind = "line", grid = True, x = "Year", y = "GDP", ylabel = "GDP", title = j)
  plt.show()
```

```python
lst = []
for j in lst:
  df_pr = df[df["Country Name"] == j]
  data = df_pr.values
  y = [0]
  for i in range(1,len(df_pr)):
    prev = data[i-1][3]
    curr = data[i][3]
    y.append(round((((curr - prev)/prev)*100),2))
  df_pr = df_pr.assign(GDP = y)
  lst.append(df_pr)
```

```python
df1 = pd.concat(lst, axis = 0)
```
